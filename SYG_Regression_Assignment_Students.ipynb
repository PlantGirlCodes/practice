### L1 Loss

```python
#L1 Absolute Error Loss
#Absolute Error for each training example 
#distance between the predicted and the actual values (doesnt matter +/-)


#related cost function: 
#Mean of these Absolute Errors (MAE):
#sum of absolute differences between predictions and actual observations
#MAE cost is more robust to outliers as compared to MSE
#so better results even when our dataset has outliers
```

### L2 Loss

```python
#L2
#Squared Error loss for each training example:
#the square of the difference between the actual and the predicted values:

#cooresponding cost function:
#MSE,mean squared error
#MSE = (sum of squared errors)/n
#MSE in this case is the cost function
#mean of the squared differences between predicted y and actual y (i.e. the residuals)

#since each error is squared, it can  magnify prediction error
#less robust to outliers


 
```

```python
def MSE_Cost_Funct(x,y,alpha=0.01,iterations=10000,tolerance=0.00001,theta_0=0,theta_1=0):
    
    #define the average - prep mean portion  
    avg = 1/(len(x))
    
    #get error and square it (total error line)
    total_error = sum([(theta_0 + (theta_1*x[i]) - y[i])**2 for i in range(len(x))])
    
    #loop to handle - case where we fail to converge
    for i in range(iterations):
        
        #set thetas to zero
        MSE_theta_0 = 0
        MSE_theta_1 = 0
        
        #loop thru data set and find dif bet ys -  update values of theta 
        for counter in range(len(x)):            
            MSE_theta_1 +=   ((theta_0 + theta_1 * x[counter]) - y[counter]) * x[counter]
        for counter in range(len(x)):            
            MSE_theta_0 +=   ((theta_0 + theta_1 * x[counter]) - y[counter])
       
        #new theta 
        new_theta_1 = theta_1 - (alpha * avg* MSE_theta_1)
        new_theta_0 = theta_0 - (alpha * avg * MSE_theta_0)
        
        #total error
        new_total_error = sum([(new_theta_0 + (new_theta_1*x[i]) - y[i])**2 for i in range(len(x))])
        
        if abs(total_error-new_total_error) <=tolerance:
            return new_theta_0,new_theta_1,i
        total_error = new_total_error
        theta_0 = new_theta_0
        theta_1 = new_theta_1
        
        return new_theta_0,new_theta_1
        
        
```
